```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
x=np.array([[200.0,17.0]])

layer_1 = keras.layers.Dense(units=3, activation="sigmoid") #this layer has 3 neurons and uses sigmoid activation function
a1 = layer_1(x)
```

```python
layer_2 = keras.layers.Dense(units=1, activation="sigmoid") #this layer has 1 neuron and uses sigmoid activation function
a2 = layer_2(a1) # pass output of previous layer as input to the next layer
```

```python
print(a1)
print(a2)
```
